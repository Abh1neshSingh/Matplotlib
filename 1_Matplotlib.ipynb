```python
x=np.linspace(0,10,100)
```

```python
x
```

```python
plt.plot() # What is plt.plot()
#Basically plt.plot helps you to make a plot
# plt.plot is nothing but is page or paper where you draw or plot anything
```

```python
plt.plot(x)
```

```python
plt.plot(x,np.sin(x))
plt.plot(x,np.cos(x))
```

```python
plt.plot(x,np.cos(x))
```

```python
plt.plot(x,np.tan(x))
```

```python
plt.plot(x,np.sin(x),'-')
```

```python
plt.plot(x,np.sin(x),'--')
```

```python
plt.plot(x,np.sin(x),'-.')
```

```python
plt.plot(x,np.sin(x),':')
```

```python
plt.figure() # make a figure size
plt.subplot(1,2,1) # it means on one page make two sub part and access the first plot
plt.plot(x,np.sin(x)) # it is ploting in first part of page
plt.subplot(1,2,2) # it means on one page make two sub part and access the Second  plot
plt.plot(x,np.cos(x))# it is ploting in Second  part of page
plt.show() # it is for showing the both plot in one page
```

```python
plt.figure() 
plt.subplot(2,2,1) 
plt.plot(x,np.sin(x)) 

plt.subplot(2,2,2) 
plt.plot(x,np.cos(x))

plt.subplot(2,2,3) 
plt.plot(x,np.tan(x))

plt.subplot(2,2,4) 
plt.plot(x,np.cos(x+1))

plt.show() 
```

```python
plt.figure() 
plt.subplot(2,1,1)   # two row one columns and access to first row
plt.plot(x,np.sin(x)) 

plt.subplot(2,1,2)    # two row one columns and access to Second  row
plt.plot(x,np.cos(x))
plt.show()
```

```python
plt.plot(x,np.sin(x),color='blue')
plt.plot(x,np.sin(x-1),color='g')
plt.plot(x,np.sin(x-2),color=(1.0,0.5,0))# rgb
```
