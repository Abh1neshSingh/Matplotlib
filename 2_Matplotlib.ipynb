```python

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# add title of plot
```

```python
plt.plot(x,np.sin(x),color='blue')
plt.ylim(-1.5,1.5)
plt.xlim(1,15)
plt.grid() # like a graph pager
plt.title("This is a (Sin) plot  before real dataset execution to understand basic syntax’s  of Matplotlib")
plt.axis('equal')
plt.show()
```

```python
# can we Provide x and y axis name 
# yes we can do that
```

```python
plt.plot(x,np.sin(x),color='blue')
plt.ylim(-1.5,1.5)
plt.xlim(1,15)
plt.grid() # like a graph pager
plt.title("This is a (Sin) plot  before real dataset execution to understand basic syntax’s  of Matplotlib")
plt.xlabel("x axis") # name of x axis
plt.ylabel("y axis") # name of y axis
plt.axis('equal')
plt.show()
```

```python
plt.plot(x,np.sin(x),color='blue', label='sinx') # label is only for give name 
plt.plot(x,np.cos(x),color='red', label='sinx')
plt.ylim(-1.5,1.5)
plt.xlim(1,15)
plt.grid() # like a graph pager
plt.title("This is a (Sin) plot  before real dataset execution to understand basic syntax’s  of Matplotlib")
plt.xlabel("x axis") # name of x axis
plt.ylabel("y axis") # name of y axis
plt.axis('equal')
plt.legend() # what ever you give label for plot it will put all the Information as well as
plt.show()
```

```python
# plot of Histogram
# Continuous data you will make histogram or distrubution plot
```

```python
data=np.random.randn(1000)
plt.hist(data) # to make histogram
plt.show()
```

```python
# by-default their are 10 bins in a histogram plot
# the smallest unit in which you are divided the data are called the bins
# bins are equal size 
```

```python
# you want to change bins size
# let see
```

```python
data=np.random.randn(1000)
plt.hist(data,bins=20) # to make histogram
plt.show()
```

```python
data=np.random.randn(1000)
plt.hist(data,bins=20,histtype='step') # to make histogram
plt.show()
```

```python
data=np.random.randn(1000)
plt.hist(data,bins=20,histtype='bar',color =(1.0,0.5,0)) # to make histogram
plt.show()
```

```python
x=np.random.rand(50)
y=np.random.rand(50)
```

```python
x
```

```python
y
```

```python
plt.figure(figsize=(6,2))
plt.scatter(x,y,c='r')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title("Scatter plot")
plt.show()
```

```python
# Scatter plot -- A scatter plot is a type of data visualization that displays individual 
#                 data points on a two-dimensional graph. 
# A pair of plot of data is a single dot which are showing in plot 
```

```python
plt.figure(figsize=(6,2))
plt.scatter(x,y,c='r',alpha=0.5)
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title("Scatter plot")
plt.show()
```

```python
plt.figure(figsize=(6,2))
plt.scatter(x,y,c='r',alpha=1,marker='*')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title("Scatter plot")
plt.show()
```

```python
# 3D plot 
```

```python
x=np.random.rand(20)
y=np.random.rand(20)
z=np.random.rand(20)

fig= plt.figure()
ax=fig.add_subplot(projection='3d') # to Visual any plot in 3d use ---add_subplot(projection='3d')
ax.scatter(x,y,z,c='#FF5733',alpha=1)
plt.xlabel("x axis")
plt.ylabel("y axis")

plt.show()
```
