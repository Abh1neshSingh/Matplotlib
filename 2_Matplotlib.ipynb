```python

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
x=np.linspace(0,10,100)
```

```python
x
```

```python
plt.plot(x,np.sin(x),color='blue')
plt.plot(x,np.sin(x-1),color='g')
plt.plot(x,np.sin(x-2),color=(1.0,0.5,0))# rgb
```

```python
# can I change limit value  of x and y axis
# yes can do it
# how
# let see
```

```python
plt.plot(x,np.sin(x),color='blue')
plt.ylim(-1.5,1.5) # let see we change it
plt.xlim(1,15)# let see we change it
```

```python
# can i have in such way plot the whole space of plot figure are fill  and no left empty spaces
# let see
```

```python
plt.plot(x,np.sin(x),color='blue')
plt.ylim(-1.5,1.5)
plt.xlim(1,15)
plt.axis('equal') # help to plot equal in page
plt.show()
```

```python
# if you want see this plot on number system or like a graph pager
# see it how to do it
```

```python
plt.plot(x,np.sin(x),color='blue')
plt.ylim(-1.5,1.5)
plt.xlim(1,15)
plt.grid() # like a graph pager 
plt.axis('equal')
plt.show()
```

```python
# add title of plot
```

```python
plt.plot(x,np.sin(x),color='blue')
plt.ylim(-1.5,1.5)
plt.xlim(1,15)
plt.grid() # like a graph pager
plt.title("This is a (Sin) plot  before real dataset execution to understand basic syntaxs  of Matplotlib")
plt.axis('equal')
plt.show()
```

```python
# can we Provide x and y axis name 
# yes we can do that
```

```python
plt.plot(x,np.sin(x),color='blue')
plt.ylim(-1.5,1.5)
plt.xlim(1,15)
plt.grid() # like a graph pager
plt.title("This is a (Sin) plot  before real dataset execution to understand basic syntaxs  of Matplotlib")
plt.xlabel("x axis") # name of x axis
plt.ylabel("y axis") # name of y axis
plt.axis('equal')
plt.show()
```

```python
plt.plot(x,np.sin(x),color='blue', label='sinx') # label is only for give name 
plt.plot(x,np.cos(x),color='red', label='sinx')
plt.ylim(-1.5,1.5)
plt.xlim(1,15)
plt.grid() # like a graph pager
plt.title("This is a (Sin) plot  before real dataset execution to understand basic syntaxs  of Matplotlib")
plt.xlabel("x axis") # name of x axis
plt.ylabel("y axis") # name of y axis
plt.axis('equal')
plt.legend() # what ever you give label for plot it will put all the Information as well as
plt.show()
```

```python
# plot of Histogram
# Continuous data you will make histogram or distrubution plot
```

```python
data=np.random.randn(1000)
plt.hist(data) # to make histogram
plt.show()
```

```python
# by-default their are 10 bins in a histogram plot
# the smallest unit in which you are divided the data are called the bins
# bins are equal size 
```

```python
# you want to change bins size
# let see
```

```python
data=np.random.randn(1000)
plt.hist(data,bins=20) # to make histogram
plt.show()
```

```python
data=np.random.randn(1000)
plt.hist(data,bins=20,histtype='step') # to make histogram
plt.show()
```

```python
data=np.random.randn(1000)
plt.hist(data,bins=20,histtype='bar',color =(1.0,0.5,0)) # to make histogram
plt.show()
```
